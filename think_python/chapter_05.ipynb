```python
from jupyturtle import forward, left, right, back, make_turtle

def draw(length):
    angle = 50
    factor = 0.6
    
    if length > 5:
        forward(length)
        left(angle)
        draw(factor * length)
        right(2 * angle)
        draw(factor * length)
        left(angle)
        back(length)

make_turtle()
draw(30)
```

```python
from jupyturtle import make_turtle

def koch(x):
    """Draws a Koch curve with the given length.
    
    Args:
        x: The length of the Koch curve segment
    """
    if x < 5:
        # Base case: draw a straight line
        forward(x)
    else:
        # Recursive case: draw 4 smaller Koch curves
        koch(x / 3)
        left(60)
        koch(x / 3)
        right(120)
        koch(x / 3)
        left(60)
        koch(x / 3)

make_turtle(delay=0)
koch(120)
```

```python
from jupyturtle import make_turtle

def draw_sierpinski(size, level):
    """Draws a Sierpiński triangle fractal.
    
    Args:
        size: The length of each side of the triangle
        level: The recursion depth (number of levels)
    """
    if level == 0:
        for _ in range(3):
            forward(size)
            left(120)
    else:
        draw_sierpinski(size / 2, level - 1)
        
        forward(size / 2)
        draw_sierpinski(size / 2, level - 1)
        
        left(120)
        forward(size / 2)
        right(120)
        draw_sierpinski(size / 2, level - 1)
        
        right(120)
        forward(size / 2)
        left(120)

make_turtle(delay=0, height=200)
draw_sierpinski(100, 3)
```
