```python
# Generator expression
large_data = (x * x for x in range(10**6))

print(next(large_data))
print(next(large_data))

for value in large_data:
    if value > 100:
        break
    print(value)
```

```python
# Define a simple decorator
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")

    return wrapper


# Use the decorator
@my_decorator
def say_hello():
    print("Hello!")


# Call the decorated function
say_hello()
```

```python
import functools
import itertools
from collections import Counter, defaultdict

# itertools common pattern: Cartesian product
colors = ["red", "green"]
sizes = ["S", "M", "L"]
product = list(itertools.product(colors, sizes))
print("Cartesian product:", product)

# functools common pattern: reduce
numbers = [1, 2, 3, 4]
sum_of_numbers = functools.reduce(lambda x, y: x + y, numbers)
print("Sum of numbers using functools.reduce:", sum_of_numbers)

# defaultdict from collections
dd = defaultdict(int)
dd["a"] += 1
dd["b"] += 2
print("Defaultdict example:", dict(dd))

# Counter from collections
counter = Counter(["apple", "banana", "apple", "orange", "banana", "apple"])
print("Counter example:", counter)

# Walrus operator (:=)
n = 10
if (half := n // 2) > 4:
    print("Half is greater than 4:", half)


# Add type hints to function signatures
def greet(name: str, age: int) -> str:
    return f"Hello, my name is {name} and I am {age} years old."


print(greet("Alice", 25))
```
