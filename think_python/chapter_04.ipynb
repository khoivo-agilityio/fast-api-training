```python
from jupyturtle import make_turtle, forward, left

make_turtle()

def square(length):
  for i in range(4):
    forward(length)
    left(90)

square(40)
square(60)
```

```python

make_turtle()

def polygon(n,length):
  angle = 360/n
  for i in range(n):
    forward(length)
    left(angle)

polygon(5, 50)
polygon(7, 60)
```

```python
import math

make_turtle(delay = 0.02)
def circle(radius):
  circumference = 2 * math.pi * radius
  n = 30
  length = circumference / n
  polygon(n, length)

circle(10)

```

```python
def polyline(n, length, angle):
  """Draws a polyline with n segments.

  Args:
      n (int): number of segments
      length (int): length of each segment
      angle (int): angle between segments
  """
  for i in range(n):
    forward(length)
    left(angle)

def polygon(n, length):
  angle = 360 / n
  polyline(n, length, angle)

def arc(radius, angle):
  arc_length = 2 * math.pi * radius * angle /360
  n = 30
  length = arc_length / n
  step_angle = angle / n
  polyline(n, length, step_angle)


def circle(radius):
  arc(radius, 360)

make_turtle(delay=0)
polygon(n=20, length=9)
arc(radius=70, angle=70)
circle(radius=10)
```

```python
from jupyturtle import penup, pendown

make_turtle()

def jump(length):
    """Move forward length units without leaving a trail.
    
    Postcondition: Leaves the pen down.
    """
    penup()
    forward(length)
    pendown()

jump(100)
```

```python
make_turtle()
def rectangle(width, height):
  for i in range(2):
    forward(width)
    left(90)
    forward(height)
    left(90)

rectangle(100, 50)
```

```python
make_turtle()

def rhombus(length):
  for i in range(2):
    forward(length)
    left(60)
    forward(length)
    left(120)

rhombus(100)

```

```python
make_turtle()

def parallelogram(width,height, angle):
  for i in range(2):
    forward(width)
    left(angle)
    forward(height)
    left(180 - angle)

parallelogram(100, 50,90)
```

```python
from jupyturtle import right

def triangle(length,angle):
  length_2 = 2 * length * math.sin(math.radians(angle/2) )
  forward(length)
  left(180-(180-angle)/2)
  forward(length_2)
  left(180-(180-angle)/2)
  forward(length)
  left(180-angle)

def shape(length, n):
  for i in range(n):
    triangle(length, 360/n)
    right(360/n)

make_turtle()
shape(50, 7)

```
