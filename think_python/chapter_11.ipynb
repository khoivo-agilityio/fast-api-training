```python
def mean(*args):
    return sum(args) / len(args)


def trimmed_mean(*args):
    low, high = min_max(args)
    trimmed = list(args)
    trimmed.remove(low)
    trimmed.remove(high)
    return mean(*trimmed)


print(mean(1, 2, 3, 4, 5))
```

```python
def sum_and_product(nums):
    total = sum(nums)
    product = 1
    for n in nums:
        product *= n
    return total, product


print(sum_and_product([1, 2, 3, 4]))
```

```python
def find_min_max(t):
    if not t:
        return None, None
    min_val = max_val = t[0]
    for item in t[1:]:
        if item < min_val:
            min_val = item
        elif item > max_val:
            max_val = item
    return min_val, max_val


print(find_min_max([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]))
```

```python
from collections import defaultdict


def is_metathesis_pair(word1, word2):
    diffs = [i for i, (a, b) in enumerate(zip(word1, word2)) if a != b]
    return (
        len(diffs) == 2
        and word1[diffs[0]] == word2[diffs[1]]
        and word1[diffs[1]] == word2[diffs[0]]
    )


word_list = [
    "converse",
    "conserve",
    "listen",
    "silent",
    "enlist",
    "inlets",
    "rescue",
    "secure",
    "recuse",
]

anagram_groups = defaultdict(list)
for word in word_list:
    anagram_groups[tuple(sorted(word))].append(word)

metathesis_pairs = []
for group in anagram_groups.values():
    for i in range(len(group)):
        for j in range(i + 1, len(group)):
            if is_metathesis_pair(group[i], group[j]):
                metathesis_pairs.append((group[i], group[j]))

print("Metathesis pairs:")
for pair in metathesis_pairs:
    print(pair)
```
