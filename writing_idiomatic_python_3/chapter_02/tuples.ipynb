```python
# Use collections.namedtuple to make tuple-heavy code more clear
from collections import namedtuple

Point = namedtuple('Point', ['x', 'y'])

point1 = Point(10, 20)
point2 = Point(30, 40)

print(f"Point 1: x={point1.x}, y={point1.y}")
print(f"Point 2: x={point2.x}, y={point2.y}")

```

```python
# Use _ as a placeholder for data in a tuple that should be ignored
x1, _ = point1
_, y2 = point2

print(f"x1: {x1}")
print(f"y2: {y2}")
```

```python
# Use tuples to unpack data
x1, y1 = point1
x2, y2 = point2

print(f"Point 1 unpacked: x1={x1}, y1={y1}")
print(f"Point 2 unpacked: x2={x2}, y2={y2}")
```

```python
# Use a tuple to return multiple values from a function
def get_midpoint(p1, p2):
  midpoint_x = (p1.x + p2.x) / 2
  midpoint_y = (p1.y + p2.y) / 2
  return (midpoint_x, midpoint_y)

midpoint = get_midpoint(point1, point2)
print(f"Midpoint: x={midpoint[0]}, y={midpoint[1]}")
```
