```python
# Avoid using a mutable object as the default value for a function argument
def append_to_list(value, my_list=None):
  if my_list is None:  
    my_list = []
  my_list.append(value)
  return my_list

result1 = append_to_list(1)
print(result1) 

result2 = append_to_list(2)
print(result2)  
```

```python
# Use return to evaluate expressions in addition to return values
def square_and_check_even(number):
  square = number ** 2
  is_even = (square % 2 == 0)
  return square, is_even

square1, is_even1 = square_and_check_even(3)
print(f"Square: {square1}, Is Even: {is_even1}")

square2, is_even2 = square_and_check_even(4)
print(f"Square: {square2}, Is Even: {is_even2}")
```

```python
# Function that demonstrates the use of keyword arguments
def introduce_yourself(name, age, city="Unknown"):
  print(f"My name is {name}, I am {age} years old, and I live in {city}.")

introduce_yourself(name="Alice", age=25, city="New York")
introduce_yourself(age=30, name="Bob") 
```

```python
# Function that accepts arbitrary arguments
def process_arguments(*args, **kwargs):
  print("Positional arguments:", args)
  print("Keyword arguments:", kwargs)

process_arguments(1, 2, 3, name="Alice", age=25)
```

```python
def double(x):
  return x * 2

def apply_function(func, value):
  return func(value)

# Treating the 'double' function as a value
result = apply_function(double, 5)
print(f"Result of applying 'double' function: {result}")
```
