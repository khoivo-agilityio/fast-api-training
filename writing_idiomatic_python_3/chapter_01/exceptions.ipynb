```python
def divide_numbers(a, b):
  try:
    result = a / b
  except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
    return None
  except TypeError:
    print("Error: Both inputs must be numbers.")
    return None
  else:
    return result

num1 = 10
num2 = 0

print("Attempting to divide...")
result = divide_numbers(num1, num2)
if result is not None:
  print(f"Result: {result}")
else:
  print("Division failed.")
```

```python
# Use Exceptions to write code in an “EAFP” style
data = {'key1': 'value1', 'key2': 'value2'}

try:
  value = data['key3'] 
  print(f"Value: {value}")
except KeyError:
  print("Error: Key not found.")
```

```python
# Avoid “swallowing” useful exceptions with bare except clauses

try:
  value = data['key3']
  print(f"Value: {value}")
except KeyError as e:
  print(f"Error: Key not found. Details: {e}")
except Exception as e:
  print(f"An unexpected error occurred: {e}")
```

```python
class CustomException(Exception):
  """Custom exception for specific error handling."""
  pass

def check_positive_number(value):
  if value <= 0:
    raise CustomException("The value must be a positive number.")
  return value

try:
  num = -5  # Example input
  print(f"Checking number: {num}")
  check_positive_number(num)
except CustomException as e:
  print(f"Custom Exception Caught: {e}")
```
