```python
def polyline(n, length, angle):
    """Draws a polyline with n segments.

    Args:
        n (int): number of segments
        length (int): length of each segment
        angle (int): angle between segments
    """
    for i in range(n):
        turtle.forward(length)
        turtle.left(angle)


def polygon(n, length):
    angle = 360 / n
    polyline(n, length, angle)


def arc(radius, angle):
    arc_length = 2 * math.pi * radius * angle / 360
    n = 30
    length = arc_length / n
    step_angle = angle / n
    polyline(n, length, step_angle)


def circle(radius):
    arc(radius, 360)


make_turtle(delay=0)
polygon(n=20, length=9)
arc(radius=70, angle=70)
circle(radius=10)
```

```python
from jupyturtle import make_turtle


def polygon(n, length):
    angle = 360 / n
    for i in range(n):
        turtle.forward(length)
        turtle.left(angle)


def circle(radius):
    circumference = 2 * math.pi * radius
    n = 30
    length = circumference / n
    polygon(n, length)


def spiral(iterations=100, initial_length=1, angle=5, growth_rate=0.02):
    """Draws a spiral by moving forward and turning repeatedly.

    Args:
        iterations: Number of steps in the spiral
        initial_length: Starting length of each segment
        angle: Angle to turn after each segment (in degrees)
        growth_rate: How much to increase segment length each step
    """
    length = initial_length
    for i in range(iterations):
        turtle.forward(length)
        turtle.left(angle)
        length = length + growth_rate  # Gradually increase segment length


make_turtle(delay=0)
spiral(iterations=150, initial_length=2, angle=10, growth_rate=0.05)
```

```python
from jupyturtle import pendown, penup  # type: ignore

make_turtle()


def jump(length):
    """Move forward length units without leaving a trail.

    Postcondition: Leaves the pen down.
    """
    penup()
    turtle.forward(length)
    pendown()


jump(100)
```

```python
make_turtle()


def parallelogram(width, height, angle):
    for i in range(2):
        turtle.forward(width)
        turtle.left(angle)
        turtle.forward(height)
        turtle.left(180 - angle)


parallelogram(100, 50, 90)
```

```python
def triangle(length, angle):
    length_2 = 2 * length * math.sin(math.radians(angle / 2))
    turtle.forward(length)
    turtle.left(180 - (180 - angle) / 2)
    turtle.forward(length_2)
    turtle.left(180 - (180 - angle) / 2)
    turtle.forward(length)
    turtle.left(180 - angle)


def shape(length, n):
    for i in range(n):
        triangle(length, 360 / n)
        turtle.right(360 / n)


make_turtle()
shape(50, 7)
```

```python
def arc(radius, angle):
    """Draws an arc with the given radius and angle."""
    arc_length = 2 * math.pi * radius * angle / 360
    n = 30
    length = arc_length / n
    step_angle = angle / n
    polyline(n, length, step_angle)


def petal(radius, angle):
    """Draws one flower petal using two arcs.

    Args:
        radius: The radius of the arcs that form the petal
        angle: The angle of each arc (controls petal width)
    """
    arc(radius, angle)
    turtle.left(180 - angle)
    arc(radius, angle)
    turtle.left(180 - angle)


def flower(radius, angle, num_petals):
    """Draws a flower with multiple petals."""
    for i in range(num_petals):
        petal(radius, angle)
        turtle.left(360 / num_petals)


make_turtle(delay=0)
# petal(radius=70, angle=60)
flower(radius=70, angle=60, num_petals=10)
```
