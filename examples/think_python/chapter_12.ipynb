```python
filename = "dr_jekyll.txt"

unique_words = {}
for line in open(filename):
    seq = line.split()
    for word in seq:
        unique_words[word] = 1

len(unique_words)

sorted(unique_words, key=len)[-5:]
```

```python
def split_line(line):
    return line.replace("—", " ").split()


split_line("coolness—frightened")
```

```python
import unicodedata

unicodedata.category("A")
```

```python
punc_marks = {}
for line in open(filename):
    for char in line:
        category = unicodedata.category(char)
        if category.startswith("P"):
            punc_marks[char] = 1


punctuation = "".join(punc_marks)
print(punctuation)
```

```python
def clean_word(word):
    return word.strip(punctuation).lower()


clean_word("“Behold!”")
```

```python
unique_words2 = {}
for line in open(filename):
    for word in split_line(line):
        word = clean_word(word)
        unique_words2[word] = 1

len(unique_words2)

sorted(unique_words2, key=len)[-5:]
```

```python
from collections import Counter

word_counter = Counter()
for line in open(filename):
    for word in split_line(line):
        word = clean_word(word)
        word_counter[word] += 1

for word, freq in word_counter.most_common(5):
    print(freq, word, sep="\t")
```

```python
def count_trigram(filename):
    trigram_counter = {}
    prev_words = []
    with open(filename) as f:
        for line in f:
            words = [clean_word(w) for w in split_line(line) if clean_word(w)]
            for word in words:
                prev_words.append(word)
                if len(prev_words) == 3:
                    trigram = tuple(prev_words)
                    trigram_counter[trigram] = trigram_counter.get(trigram, 0) + 1
                    prev_words.pop(0)
    return trigram_counter


def process_word_trigram(trigram_counter, n=5):
    items = sorted(trigram_counter.items(), key=lambda x: x[1], reverse=True)
    for trigram, freq in items[:n]:
        print(freq, " ".join(trigram), sep="\t")


trigram_counter = count_trigram(filename)
process_word_trigram(trigram_counter)
```
