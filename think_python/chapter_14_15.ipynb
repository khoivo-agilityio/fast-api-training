```python
from datetime import datetime


class Time:
    def __init__(self, hour, minute, second):
        self.hour = hour
        self.minute = minute
        self.second = second

    def time_to_int(self):
        minutes = self.hour * 60 + self.minute
        seconds = minutes * 60 + self.second
        return seconds

    def __add__(self, other):
        seconds = self.time_to_int() + other.time_to_int()
        return Time.int_to_time(seconds)

    @staticmethod
    def int_to_time(seconds):
        minutes, second = divmod(seconds, 60)
        hour, minute = divmod(minutes, 60)
        return Time(hour, minute, second)

    def __str__(self):
        s = f"{self.hour:02d}:{self.minute:02d}:{self.second:02d}"
        return s


def subtract_time(time1, time2):
    """
    Subtracts two Time objects and returns the interval in seconds.
    Assumes both times are on the same day.
    """
    t1 = datetime(1, 1, 1, time1.hour, time1.minute, time1.second)
    t2 = datetime(1, 1, 1, time2.hour, time2.minute, time2.second)
    delta = t1 - t2
    return abs(delta.total_seconds())


time1 = Time(14, 30, 0)
time2 = Time(12, 15, 0)
interval = subtract_time(time1, time2)
print(f"The interval between the two times is {interval} seconds.")


def is_after(t1, t2):
    """Checks whether `t1` is after `t2`."""
    dt1 = datetime(1, 1, 1, t1.hour, t1.minute, t1.second)
    dt2 = datetime(1, 1, 1, t2.hour, t2.minute, t2.second)
    return dt1 > dt2


print(is_after(time1, time2))


class Date:
    """Represents a year, month, and day"""

    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day


def make_date(year, month, day):
    """Creates and returns a Date object."""
    return Date(year, month, day)


def print_date(date):
    """Prints a Date object in YYYY-MM-DD format."""
    print(f"{date.year:04d}-{date.month:02d}-{date.day:02d}")


def date_to_tuple(date):
    """Converts a Date object to a tuple (year, month, day)."""
    return (date.year, date.month, date.day)


def is_after(date1, date2):
    """Checks if date1 is after date2."""
    return date_to_tuple(date1) > date_to_tuple(date2)


date1 = make_date(2025, 6, 22)
date2 = make_date(2026, 9, 17)

print_date(date1)


start = Time(9, 45, 0)
duration = Time(1, 32, 0)
end = start + duration
print(end)
```

```python
class Date:
    """Represents a year, month, and day."""

    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day

    def __str__(self):
        """Returns a string representation of the date in YYYY-MM-DD format."""
        return f"{self.year:04d}-{self.month:02d}-{self.day:02d}"

    def to_tuple(self):
        """Returns a tuple representation of the date."""
        return (self.year, self.month, self.day)

    def is_after(self, other):
        """Returns True if this date is after the other date."""
        return self.to_tuple() > other.to_tuple()


date1 = Date(1933, 6, 22)
print(date1)

date2 = Date(1933, 9, 17)
print(date2.is_after(date1))
```
