```python
# Symmetric Difference: Elements in either set_a or set_b, but not in both
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

union_result = set_a.union(set_b)
print("Union:", union_result)

intersection_result = set_a.intersection(set_b)
print("Intersection:", intersection_result)

difference_result = set_a.difference(set_b)
print("Difference (A - B):", difference_result)

symmetric_difference_result = set_a.symmetric_difference(set_b)
print("Symmetric Difference:", symmetric_difference_result)
```

```python
# Use a set comprehension to generate sets concisely
squares_set_a = {x**2 for x in set_a}
print("Squares of elements in set_a:", squares_set_a)

greater_than_4_set_b = {x for x in set_b if x > 4}
print("Elements in set_b greater than 4:", greater_than_4_set_b)
```

```python
# Use sets to determine if two lists share any common values
list1 = [1, 2, 3, 7]
list2 = [5, 6, 7, 8]

set1 = set(list1)
set2 = set(list2)

common_values = set1.intersection(set2)
if common_values:
  print("Common values:", common_values)
else:
  print("No common values.")
```

```python
# Use sets to eliminate duplicate entries from Iterable containers
unique_list1 = list(set(list1))
unique_list2 = list(set(list2))

print("Unique elements in list1:", unique_list1)
print("Unique elements in list2:", unique_list2)
```
