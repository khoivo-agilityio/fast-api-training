```python
# Use if statements to return early on error conditions
def process_data(data):
    if not data:
        return "Error: No data provided"

    if not isinstance(data, list):
        return "Error: Data must be a list"

    if not all(isinstance(item, int) for item in data):
        return "Error: All items in the data must be integers"

    return [item * 2 for item in data]


result = process_data([1, 2, 3])
print(result)

error_result = process_data("")
print(error_result)
```

```python
# Chain comparisons to make if statements more concise
def is_within_range(number, lower, upper):
    return lower <= number <= upper


print(is_within_range(5, 1, 10))
print(is_within_range(15, 1, 10))
```

```python
# Avoid placing conditional branch code on the same line as the colon
def check_positive(number):
    if number > 0:
        return "Positive"
    else:
        return "Non-positive"


print(check_positive(10))
print(check_positive(-5))
```

```python
# Avoid repeating variable name in compound if statement
def is_valid_result(data):
    if data and isinstance(data, list):
        return "Valid result"
    return "Invalid result"


print(is_valid_result(result))
print(is_valid_result(error_result))
```

```python
# Avoid comparing directly to True, False, or None
if result:
    print("Result is truthy")
else:
    print("Result is falsy")

if error_result:
    print("Error result is truthy")
else:
    print("Error result is falsy")
```
