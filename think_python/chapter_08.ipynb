```python
def head_like(input_file, num_lines, output_file=None):
    infile = open(input_file, "r")
    lines = []

    for _ in range(num_lines):
        line = infile.readline()
        if line == "":
            break
        lines.append(line)

    infile.close()

    if output_file is None:
        for line in lines:
            print(line, end="")
    else:
        outfile = open(output_file, "w")
        for line in lines:
            outfile.write(line)
        outfile.close()

# Print first 5 lines to the screen
head_like("pg345.txt", 5)

# Write first 10 lines to another file
head_like("pg345.txt", 10, "output.txt")
```

```python
def uses_any(word, letters):
    for letter in word:
        if letter in letters:
            return True
    return False


def uses_only(word, letters):
    for letter in word:
        if letter not in letters:
            return False
    return True
  
def has_all(word, letters):
    for letter in letters:
        if letter not in word:
            return False
    return True

def check_word(word, required="", forbidden="", bad_positions=None):
  """Check if a word meets certain criteria.

  Args:
      word (str): The candidate 5-letter word.
      required (str, optional): Letters that must appear in the word. Defaults to "".
      forbidden (str, optional): Letters that must NOT appear in the word. Defaults to "".
      bad_positions (dict, optional): Mapping of letter -> list of forbidden indexes.
                   e.g. {'e': [2, 4]}. Defaults to None. If None, no position restrictions are applied.

  Returns:
      bool: True if the word meets all criteria, False otherwise.
  """

  if bad_positions is None:
    bad_positions = {}

  if len(word) != 5:
    return False

  if uses_any(word, forbidden):
    return False

  if not has_all(word, required):
    return False

  for letter, positions in bad_positions.items():
    for pos in positions:
      if word[pos] == letter:
        return False

  return True

required = "e"          
forbidden = "spadclrk"  

bad_positions = {
    'e': {2, 4}         
}

word_list = ["enemy", "entry", "eject", "evoke", "excel", "elope", "exert"]

def solve_wordle(word_list, required, forbidden, bad_positions):
    possible = []

    for word in word_list:
        if check_word(word, required, forbidden, bad_positions):
            possible.append(word)

    return possible


possible_words = solve_wordle(word_list, required, forbidden, bad_positions)

print(len(possible_words))
print(possible_words[:10])  

```
