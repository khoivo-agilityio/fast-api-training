```python
def repeat(world, n):
  print(world * n)
spam = 'Spam, '

def first_two_lines():
  repeat('Spam, ', 4)
  repeat('Spam, ', 4)
def last_three_lines():
  repeat('Spam, ', 2)
  print('(Lovely Spam, Wonderful Spam!)')
  repeat('Spam, ', 2)

def print_song():
  first_two_lines()
  last_three_lines()

print_song()



```

```python
char_1 = "Monty"
char_2 = "Python's"
char_3 = "Flying Circus"

def print_right(char):
  space = ''
  for i in range(40-len(char)):
    space += ' '
  print(space + char)

print_right(char_1)
print_right(char_2)
print_right(char_3)
```

```python
def triangle(char, n):
  for i in range(n):
    print(char * (i+1))

triangle('*', 5)

```

```python
def rectangle(char, width, height):
  for i in range(height):
    print(char * width)

rectangle('*', 5, 3)

```
