```python
def fibonacci(n):
    if n == 0:
        return 0
    elif  n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(10))
```

```python
def factorial(n):
    if not isinstance(n, int):
        print('factorial is only defined for integers.')
        return None
    elif n < 0:
        print('factorial is not defined for negative numbers.')
        return None
    elif n == 0:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))
```

```python
import math

def distance(x1, y1, x2, y2):
    """Computes the distance between two points.
    
    Args:
        x1, y1: Coordinates of the first point
        x2, y2: Coordinates of the second point
    
    Returns:
        The Euclidean distance between the two points
    """
    dx = x2 - x1
    dy = y2 - y1
    return math.sqrt(dx**2 + dy**2)

print(distance(1, 2, 4, 6)) 
print(distance(0, 0, 3, 4)) 
print(distance(0, 0, 0, 0)) 
```

```python
def hypot(a, b):
    """Returns the length of the hypotenuse of a right triangle.
    
    Args:
        a: Length of the first leg
        b: Length of the second leg
    
    Returns:
        The length of the hypotenuse
    """
    return math.sqrt(a**2 + b**2)

hypot(3, 4) 
hypot(5, 12) 
hypot(1, 1) 
```

```python
def is_between(x,y,z):
    return x <= y and y <= z

print(is_between(1,2,3))
print(is_between(1,3,2))
print(is_between(2,1,3))
```

```python
def ackermann (m, n):
  """Computes the Ackermann function.

  Args:
      m (int): The first argument.
      n (int): The second argument.

  Returns:
      int: The result of the Ackermann function.
  """
  if m == 0:
    return n + 1
  elif m > 0 and n == 0:
    return ackermann(m - 1, 1)
  elif m> 0 and n > 0:
    return ackermann(m - 1, ackermann(m, n - 1))
      
print (ackermann(3, 5))
```

```python
def gcd(a, b):
  """Computes the greatest common divisor (GCD) of two numbers.

  Args:
      a (int): The first number.
      b (int): The second number.

  Returns:
      int: The GCD of a and b.
  """
  while b != 0:
    a, b = b, a % b
  return a

print(gcd(48, 18))
```
