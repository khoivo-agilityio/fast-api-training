```python
# Use a list comprehension to create a transformed version of an existing list
numbers = [1, 2, 3, 4, 5]

squared_numbers = [x**2 for x in numbers]

print(squared_numbers)
```

```python
# Make use of negative indexes
last_number = numbers[-1]

second_last_squared = squared_numbers[-2]

print(f"Last number: {last_number}")
print(f"Second-to-last squared number: {second_last_squared}")
```

```python
# Use a list comprehension to filter even numbers and square them
even_squared = [x**2 for x in numbers if x % 2 == 0]

print(even_squared)
```

```python
total_sum = sum(numbers)
print(f"Sum of numbers: {total_sum}")
```

```python
all_numbers_positive = all(x > 0 for x in numbers)
print(f"Are all numbers positive? {all_numbers_positive}")
```

```python
# Use the * operator to represent the “rest” of a list
first, *rest = numbers

print(f"First number: {first}")
print(f"Rest of the numbers: {rest}")
```
