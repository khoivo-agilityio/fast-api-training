```python
from doctest import run_docstring_examples

file_object = open('words.txt')

line = file_object.readline()
word = line.strip()

def has_e(word):    
  return 'e' in word.lower()

def uses_any(word, letters):
  """
  Checks if the word uses any of the specified letters.

  Args:
      word (str): The word to check.
      letters (str): The letters to look for.

  Returns:
      bool: True if the word uses any of the letters, False otherwise.
  """
  for letter in word.lower():
    if letter in letters.lower():
      return True
    return False

total = 0
count = 0
for line in open('words.txt'):
    word = line.strip()
    total += 1
    if has_e(word):
        count += 1
    
print(f'There are {total} words in the file.')
print(f'There are {count} words with an "e" in them.')

def run_doctests(func):
  run_docstring_examples(func, globals(), name=func.__name__)

run_doctests(uses_any)
```

```python
def uses_none(word, forbidden):
  """Checks whether a word avoids forbidden letters.
  
  >>> uses_none('banana', 'xyz')
  True
  >>> uses_none('apple', 'efg')
  False
  >>> uses_none('grape', 'xyz')
  True
  """
  for letter in word.lower():
    if letter in forbidden.lower():
      return False
  return True

print(uses_none('banana', 'xyz'))  
print(uses_none('apple', 'efg'))   

run_doctests(uses_none)
```

```python
def uses_only(word, available):
  """Checks whether a word uses only the available letters.
  
  >>> uses_only('banana', 'ban')
  True
  >>> uses_only('apple', 'apl')
  False
  >>> uses_only('dad', 'ad')
  True
  """
  for letter in word.lower():
    if letter not in available.lower():
      return False
  return True

print(uses_only('hello', 'xyz'))
print(uses_only('hello', 'heloxyz'))
```

```python
def check_word(word, available, required):
  """Check whether a word is acceptable.
  
  >>> check_word('color', 'ACDLORT', 'R')
  True
  >>> check_word('ratatat', 'ACDLORT', 'R')
  True
  >>> check_word('rat', 'ACDLORT', 'R')
  False
  >>> check_word('told', 'ACDLORT', 'R')
  False
  >>> check_word('bee', 'ACDLORT', 'R')
  False
  """
  word = word.lower()
  available = available.lower()
  required = required.lower()
  if len(word) < 4:
    return False
  if required not in word:
    return False
  if not uses_only(word, available):
    return False
  return True

def word_score(word, available):
  """Compute the score for an acceptable word.
  
  >>> word_score('card', 'ACDLORT')
  1
  >>> word_score('color', 'ACDLORT')
  5
  >>> word_score('cartload', 'ACDLORT')
  15
  """
  word = word.lower()
  available = available.lower()
  if len(word) == 4:
    return 1
  score = len(word)
  if set(available) <= set(word):
    score += 7  
  return score

print(check_word('color', 'ACDLORT', 'R'))
print(check_word('ratatat', 'ACDLORT', 'R'))
```

```python
def uses_all(word, required):
  """Checks whether a word uses all the required letters.
  
  >>> uses_all('banana', 'ab')
  True
  >>> uses_all('apple', 'ae')
  False
  >>> uses_all('grape', 'gp')
  True
  """
  for letter in required.lower():
    if not uses_only(letter, word):
      return False
  return True

print(uses_all('banana', 'ab'))  
print(uses_all('apple', 'ae'))   
```

```python
def uses_all(s1, s2):
  """Checks if all characters in s2 are in s1, allowing repeats."""
  for char in s2:
    if not uses_any(s1, char):
      return False
  return True


print(uses_all('banana', 'ab'))   
print(uses_all('apple', 'ae'))    
print(uses_all('grape', 'gp'))    
run_doctests(uses_all)
```
