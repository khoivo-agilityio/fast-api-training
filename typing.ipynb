```python
from typing import (
    Callable,
    Generic,
    Literal,
    Optional,
    TypeAlias,
    TypedDict,
    TypeVar,
    Union,
)


# Basic Types
def add(a: int, b: int) -> int:
    return a + b


# Union & Optional
def greet(name: Optional[str]) -> str:
    return f"Hello {name or 'World'}"


# Type Aliases
Numbers: TypeAlias = list[int]
Mapping: TypeAlias = dict[str, float]

values: Numbers = [1, 2, 3]
scores: Mapping = {"math": 9.5, "physics": 8.0}

# Generics (TypeVar)
T = TypeVar("T")


def first(items: list[T]) -> T:
    return items[0]


# Generics (Generic class)
class Box(Generic[T]):
    def __init__(self, value: T):
        self.value = value

    def get(self) -> T:
        return self.value


# Callables
Processor = Callable[[int, int], int]


def apply(op: Processor, x: int, y: int) -> int:
    return op(x, y)


# Union
Result = Union[int, str]


def stringify(value: Result) -> str:
    return str(value)


# Literal Types
Status = Literal["pending", "success", "failed"]


# TypedDict
class User(TypedDict):
    id: int
    name: str
    status: Status


user: User = {
    "id": 1,
    "name": "Khoi",
    "status": "success",
}
```
