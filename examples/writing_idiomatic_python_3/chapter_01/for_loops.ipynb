```python
fruits = ['apple', 'banana', 'cherry', 'date']

# Using enumerate to loop with index and value
for index, fruit in enumerate(fruits):
  print(f"Index {index}: {fruit}")
```

```python
# Iterating over the list of fruits
for fruit in fruits:
  print(f"I like {fruit}")
```
