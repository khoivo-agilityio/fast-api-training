```python
def mean(*args):
    return sum(args) / len(args)
  
def trimmed_mean(*args):
    low, high = min_max(args)
    trimmed = list(args)
    trimmed.remove(low)
    trimmed.remove(high)
    return mean(*trimmed)

print(mean(1,2,3,4,5))
```
