```python
with open("words.txt") as f:
    word_list = [line.strip() for line in f]
    
def is_anagram(word1, word2):
  return sorted(word1) == sorted(word2)

anagrams_of_takes = [word for word in word_list if is_anagram(word, "takes") and word != "takes"]
anagrams_of_takes
```

```python
def is_palindrome(word):
  return word == ''.join(reversed(word))

for word in word_list:
  if len(word) >= 7 and is_palindrome(word):
    print(word)
```

```python
def reverse_sentence(sentence):
  words = sentence.split()
  if not words:
    return ""
  reversed_words = words[::-1]
  
  reversed_words = [reversed_words[0].capitalize()] + [w.lower() for w in reversed_words[1:]]
  return ' '.join(reversed_words)

print(reverse_sentence("Hello world from Python"))
```
